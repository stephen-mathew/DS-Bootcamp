### Manipulations on Performance data

```python
dfPerformance['Date']=pd.to_datetime(dfPerformance['Date'])
```

```python
# str(dfPerformance['Date'].dt.week) + "-" + str(dfPerformance['Date'])

dfPerformance['Week']=dfPerformance['Date'].dt.isocalendar().week
```

```python
dfPerformance['Average Response Time D']=pd.to_datetime(dfPerformance['Average Response Time'],format='%H:%M:%S')
```

```python
dfPerformance['Average Resolution Time D']=pd.to_datetime(dfPerformance['Average Resolution Time'],format='%H:%M:%S')
```

```python
dfPerformance['Average Response Time D']=(dfPerformance['Average Response Time D']-datetime(1900,1,1)).dt.total_seconds()
```

```python
dfPerformance['Average Resolution Time D']=(dfPerformance['Average Resolution Time D']-datetime(1900,1,1)).dt.total_seconds()
```
